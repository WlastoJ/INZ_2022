```python
import pandas as pd
import sklearn as skl
import matplotlib.pyplot as plt
import time

pd.options.mode.chained_assignment = None
pd.set_option('display.max_columns',None)
```

```python
model = pd.read_csv('model.csv', index_col='track_id')  
model.head()
```

```python
model["test"] = 0
```

```python
model.loc[model["('set', 'split')"] == "test", "test"] = 1
```

```python
model=model.drop(columns=["('set', 'split')"])
```

```python
modelfinal=pd.get_dummies(model)
modelfinal.head()
```

```python
modelFolk = modelfinal
```

```python
modelFolk=modelFolk.drop(columns="genre_International")
modelFolk=modelFolk.drop(columns='genre_Jazz')
modelFolk=modelFolk.drop(columns='genre_Rock')
modelFolk=modelFolk.drop(columns='genre_Experimental')
modelFolk=modelFolk.drop(columns='genre_Electronic')
modelFolk=modelFolk.drop(columns='genre_Hip-Hop')
modelFolk=modelFolk.drop(columns='genre_Instrumental')
modelFolk=modelFolk.drop(columns='genre_Blues')
modelFolk=modelFolk.drop(columns='genre_Classical')
modelFolk=modelFolk.drop(columns='genre_Old-Time / Historic')
modelFolk=modelFolk.drop(columns='genre_Pop')
```

```python
modelFolk.head()
```

```python
modelFolk.to_csv('folk.csv')
```

```python
TRAINmodelFolk=modelFolk.loc[modelFolk["test"] == 0]
TRAINmodelFolk=TRAINmodelFolk.drop(columns="test")
TESTmodelFolk=modelFolk.loc[modelFolk["test"] == 1]
TESTmodelFolk=TESTmodelFolk.drop(columns="test")
```

```python
TRAINmodelFolk.genre_Folk.value_counts()
```

```python
TESTmodelFolk.genre_Folk.value_counts()
```

```python
Xtr=TRAINmodelFolk.drop(columns="genre_Folk")
Ytr=TRAINmodelFolk["genre_Folk"]
```

```python
Xtest=TESTmodelFolk.drop(columns="genre_Folk")
Ytest=TESTmodelFolk["genre_Folk"]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn import tree
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier


from sklearn import metrics
from sklearn.metrics import precision_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix


lr=LogisticRegression(max_iter=13129, solver='sag')
rfc=RandomForestClassifier(n_estimators=100)
svc=LinearSVC(C=10,max_iter=12500,dual=False)
tree = tree.DecisionTreeClassifier()
gbc = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0, max_depth=1, random_state=0)
nn = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=1)
```

# LOGISTIC REGRESSION

```python
tic = time.perf_counter()

lr.fit(Xtr, Ytr)

toc = time.perf_counter()

print(f"Wykonanie zajelo {toc - tic:0.4f} sekund")

print(lr.score(Xtr,Ytr))
```

```python
predykcja=lr.predict(Xtest)
print("Sredni blad bezwzgledny", metrics.mean_absolute_error(Ytest,predykcja))

print("Precision score: ",precision_score(Ytest,predykcja, average='weighted',zero_division=1))

print("Accuracy score: ",accuracy_score(Ytest,predykcja, normalize=False),"/1293")

print(lr.score(Xtest,Ytest))
```

```python
confusion_matrix(Ytest,predykcja)
```
